```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
N_values = [5, 10, 15, 20, 25, 30, 100]

for N in N_values:

    # read results
    
    results = open(f"results{N}.txt", "r")
    values = results.readline().rstrip(" \n").split(" ")
    values = [eval(i) for i in values]
    
    # create X axes for graphs
    
    figs, ax = plt.subplots()
    bins = np.arange(-N - 1, N + 2, 2)
    X = np.linspace(-N - 1, N + 1, 1000)
    
    # real normal dist pdf

    normal_pdf = ((1 / math.sqrt(N) / math.sqrt(2 * math.pi)) * np.exp(-0.5 * (X / math.sqrt(N))**2))
    
    # graph pdf
    
    ax.hist(values, bins, density = True)
    plt.plot(X, normal_pdf)
    plt.title(f"pdf for N = {N}")
    plt.show()
    
    # calculate cdf for results
    
    cdf = np.zeros(N + 2, dtype=int)
    for value in values:
        for i in range((int(value) + N) // 2):
            cdf[-i - 2] += 1
    cdf = cdf / 5000
    cdf[N + 1] = 1
    
    # real normal dist cdf
    
    real_cdf = 0.5 * (1 + np.vectorize(math.erf)(X/math.sqrt(N)/math.sqrt(2)))
    
    # graph cdf
    
    plt.plot(bins, cdf)
    plt.plot(X, real_cdf)
    plt.title(f"cdf for N = {N}")
    plt.show()
    
    results.close()
```
